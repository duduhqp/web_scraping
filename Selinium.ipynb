```python
def check_verified():
    check = driver.find_element_by_xpath('//*[@title="Verificado"]')
    return check
```

```python
def get_post_date():
    date = driver.execute_script('''return document.getElementsByClassName('_1o9PC Nzb55')[0].getAttribute('datetime')''')
    return date
```

```python
def get_total_likes():
    
    try:
        likes = driver.execute_script('''return document.getElementsByClassName('Nm9Fw')[0].lastElementChild.getElementsByTagName('span')[0].innerText''')
        likes = [i.split('\n')[0] for i in likes]
        likes = ''.join(likes)
        
        return likes
    except:
        return 0
```

```python
def get_post_exit():
    exit = driver.execute_script('''return document.getElementsByClassName('wpO6b  ')''')
    tam = len(exit)
    return exit[tam-1]
```

```python
def get_next_post():
    nextPost = driver.execute_script('''return document.getElementsByClassName('_65Bje')[0]''')
    return nextPost
```

```python
#Se verificado ent√£o getElementsByTagName('span')[2]
#Se N√ÇO verificado ent√£o getElementsByTagName('span')[1]

def get_post_description():
    try:
        check = check_verified()  
        description = driver.execute_script('''return document.getElementsByClassName('C4VMK')[0].getElementsByTagName('span')[2].innerText''')
    except:  
        description = driver.execute_script('''return document.getElementsByClassName('C4VMK')[0].getElementsByTagName('span')[1].innerText''')
        
    return description 
```

```python
#scraping_profile

def scraping_profile_list(data,profile): 
    
    images = driver.execute_script('''return document.getElementsByClassName('_9AhH0')''')
    image = images[1]
    image.click()

   
    count = 0
    while (get_next_post() != None) and (count < 50):
        sleep(3)
        count += 1
        #Choosing data colected
        date = get_post_date()
        description = get_post_description()
        totalLikes = get_total_likes()

        get_next_post().click()

        #Creating and storing DataFrame
        d = {'Description':description,
             'Likes':totalLikes,
            'Date':date,
            'Profile':profile}
        data.append(d)

    get_post_exit().click()    
    
    return
```

```python
# List of profiles scraped

keyword_list = ["partagao","manubiar","chefbrugand","leomchef","teiagastronomia","marciaareiasgastronomia","soulbaker_","semmedida","acasadoporcobar","bardadonaonca","ruan.felixs","jimmy.ogro"]
```

```python
data = []

for keyword in keyword_list:
    searchbox = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Pesquisar']")))
    searchbox.clear()
    sleep(4)
    searchbox.send_keys(keyword)
    sleep(4)
    searchbox.send_keys(Keys.ENTER)
    sleep(4)
    searchbox.send_keys(Keys.ENTER)
    sleep(4)

    scraping_profile_list(data, keyword)

#Printing DataFrame
df = pd.DataFrame(data)
print(df)
    
```
